#Isolation forest

```python
from sklearn.ensemble import IsolationForest
```

```python
data = pd.read_csv("/content/iris.csv")
data_encoded = pd.get_dummies(data)
```

```python
#training the model
clf = IsolationForest(random_state=10,contamination=0.01) #contamination is based on the domain knowledge how many outliers could you expect
clf.fit(data_encoded) #n_estimators means the number of isolation trees
```

```python
#predictions:
y_pred_outliers = clf.predict(data_encoded)
```

```python
y_pred_outliers
```
