#Emotion Mining

```python
import pandas as pd
```

```python
affin = pd.read_csv("/content/Afinn.csv",sep=",",encoding="latin-1")
affin.head()
```

```python
import numpy as np
import string
import spacy
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread
from wordcloud import WordCloud
%matplotlib inline
```

```python
book = pd.read_csv("/content/apple.txt",error_bad_lines=False)
book =(x.strip() for x in book.x)
book = (x for x in book if x)
```

```python
#to apply emotion mining we neet this nltk and punkt
import nltk
nltk.download("punkt")
```

```python
from nltk import tokenize
sentenses = tokenize.sent_tokenize("".join(book))
sentenses[5:15]
#we doing this because each review is being broken down into sentences and each review is taken as a document
```

```python
set_df = pd.DataFrame(sentenses,columns=["Sentenses"])
set_df
```

```python
affinity_scores = affin.set_index("word")["value"].to_dict()
affinity_scores
```

```python
#Custom func: score each word in a sentense in the lemitized form,
#but cal the score for the whole sentense
nlp = spacy.load("en_core_web_sm")
sentiment_lexicon = affinity_scores

def calculate_sentiment(text:str=None):
  sent_score = 0
  if text:
    sentense = nlp(text)
    for word in sentense:
      sent_score += sentiment_lexicon.get(word.lemma_,0)
      return sent_score
```

```python
#test if it works
calculate_sentiment(text="amazing")
```

```python
set_df["Sentiment_value"] = set_df["Sentenses"].apply(calculate_sentiment)
```
