```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv("/content/Movie.csv")
movie.head(-5)
```

```python
#to check the number of unique users in my dataset:
len(movie.userId.unique())
```

```python
len(movie.movie.unique())
```

```python
#we will create our user-item interaction matrix:
user_movie = movie.pivot(index="userId",
                         columns="movie",
                         values="rating").reset_index(drop=True)
```

```python
user_movie
```

```python
user_movie.index = movie.userId.unique()
```

```python
user_movie
```

```python
user_movie.fillna(0,inplace=True)
```

```python
user_movie
```

```python
#calculating the cosine similarity between users:
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
#the reason we are using 1- is because we want to find the similarity and not the dissimilarity:
user_sim = 1-pairwise_distances(user_movie.values, metric="cosine")
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
user_sim_df.index = movie.userId.unique()
user_sim_df.columns = movie.userId.unique()
```
